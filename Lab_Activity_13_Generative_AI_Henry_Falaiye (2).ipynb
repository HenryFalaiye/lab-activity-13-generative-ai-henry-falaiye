```python
# Import libraries
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
model_name = "gpt2"

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Set model to evaluation mode
model.eval()
```

```python
def generate_text(prompt, max_length=100, temperature=0.7, top_k=50, top_p=0.95):
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    outputs = model.generate(
        inputs,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True,
        num_return_sequences=1
    )

    text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return text
```

```python
prompt1 = "Once upon a time"

result1 = generate_text(prompt1)
print(result1)
```

```python
prompt2 = "The future of artificial intelligence is"

short_text = generate_text(prompt2, max_length=50)
long_text = generate_text(prompt2, max_length=200)

print("Short:\n", short_text)
print("\nLong:\n", long_text)
```

```python
prompt3 = "The benefits of renewable energy include"

low_temp = generate_text(prompt3, temperature=0.2)
high_temp = generate_text(prompt3, temperature=1.5)

print("Low Temperature:\n", low_temp)
print("\nHigh Temperature:\n", high_temp)
```

```python
prompt4 = "In the year 2050, transportation will"

default_output = generate_text(prompt4)
controlled_output = generate_text(prompt4, top_k=10, top_p=0.5)

print("Default:\n", default_output)
print("\nControlled:\n", controlled_output)
```

```python
base_prompt = "Write a short story about"

topics = [
    "a robot learning emotions",
    "a lost city under the ocean",
    "a time traveler stuck in the past",
    "an AI taking over the world"
]

for topic in topics:
    full_prompt = f"{base_prompt} {topic}"
    print("\nPrompt:", full_prompt)
    print(generate_text(full_prompt))
```

```python
user_prompt = input("Enter your prompt: ")
print(generate_text(user_prompt))
```

```python
def ask_gpt2(prompt):
    return generate_text(prompt, max_length=150)
```

```python
system_prompt = "You are a helpful AI assistant."
user_prompt = "Explain machine learning in simple terms."

full_prompt = system_prompt + "\nUser: " + user_prompt + "\nAI:"

response = ask_gpt2(full_prompt)

print(response)
```

```python
def chat_with_gpt():
    print("Chat started (type 'exit' to stop)\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() == "exit":
            print("Chat ended.")
            break

        prompt = "User: " + user_input + "\nAI:"
        response = ask_gpt2(prompt)

        print("AI:", response)
```

```python
chat_with_gpt()
```

```python
blog_prompt = "Write a blog post about Python for data science."

blog = ask_gpt2(blog_prompt)

print(blog)
```

### Generative AI
Generative AI models like GPT-2 can generate human-like text based on input prompts.

### Temperature
Temperature controls randomness in text generation. Lower values produce deterministic output, while higher values increase creativity.

### Top-k and Top-p
These parameters control how the model selects words during generation.
